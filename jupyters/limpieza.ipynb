```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..')
from src import sp_limpieza as sp
```

```python
df = pd.read_csv('../data/data_transformacion_limpieza.csv')
```

```python
sp.eda_preliminar(df)
```

```python
sp.valores_minusculas(df) # Cambia los valores de las columnas a minúsculas
```

```python
df['duration'][0].split()[0] # Se observa que la columna 'duration' tiene valores con formato de tiempo
```

```python
df['duration'] = df['duration'].apply(lambda x: x.split()[0] if isinstance(x,str) else x) # Elimina los valores de la columna duration que no sean numéricos
```

```python
dolar = ['acquisition_cost','total($)'] # Columnas a las que se les quitará el signo de dolar
sp.dolares (df, dolar)
```

```python
df['acquisition_cost'].equals(df['total($)']) # Se verifica que las columnas sean iguales
df.drop('acquisition_cost', axis=1, inplace=True) # Se elimina la columna acquisition_cost ya que es igual a total($)
```

```python
coma = ['conversion_rate', 'roi', 'clicks', 'impressions', 'engagement_score'] # Columnas a las que se les quitará la coma
sp.comas(df, coma)
```

```python
df[['genre', 'ages']] = df['target_audience'].str.split(' ', expand=True) # Se separa la columna target_audience en dos columnas
```

```python
df['genre'] = df['genre'].apply(lambda x: np.nan if x == 'all' else x)  # Se cambia el valor 'all' por NaN
df['ages'] = df['ages'].apply(lambda x: 'all ages' if x == 'ages' else x)  # Se cambia el valor 'ages' por 'all ages'
df.drop('target_audience', axis=1, inplace=True) # Se elimina la columna target_audience
```

```python
sp.convertir_columnas(df, '%Y-%m-%d') # Se convierten las columnas a su tipo de dato correspondiente
```

```python
df['year'] = df['date'].dt.year # Se crea una columna con el año
df['month'] = df['date'].dt.month # Se crea una columna con el mes
df['day_name'] = df['date'].dt.day_name() # Se crea una columna con el día
df['quarter'] = df['date'].dt.quarter # Se crea una columna con el trimestre
```

```python
sp.valores_minusculas(df) # Se verifica que los valores de las columnas estén en minúsculas
```

```python
sp.quitar_espacios(df) # Se quitan los espacios en blanco de las columnas
```

```python
df.to_csv('../data/data_limpios.csv', index=False) # Se guarda el archivo con las transformaciones realizadas
```
